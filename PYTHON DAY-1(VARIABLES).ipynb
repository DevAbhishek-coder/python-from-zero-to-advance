```python
#VARIABLES:- A variable in Python is a name given to a value stored in memory.
# It works like a container that holds data which can be used and changed later.
```

```python
#EXAMPLE
 
a = 10; 
print(a)

#here a is variable name and 10 is value stored in memory

name = "Abhishek"
print(name)
# Here name is a variable and Abhishek is a value stored in a memory


#We can check the data type of variables

print(type(a)) #type is a inbuilt function in python, which can give the data type of a variables.
```

```python
#  #There is a one case, when we giving same variable name which is stored in a memory
# and now changing it's data value. when we run the code, python will override the previous data value
# and return new given data value. 
```

```python
#Example
print(a) 

a = 20 #here we have changed the data value and once we run the program, we got 20 insted of 10, here 
# python override the previous data.
print(a)
```

```python
# #The old value still exists in memory for a short time, but no variable references it.

# This phase is called:

# "Unreferenced Object"
```
